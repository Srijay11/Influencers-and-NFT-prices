```python

# Create API object
api = tweepy.API(auth)

# Define the hashtag to search for
hashtag = "#Nfts"

# Define the time interval to search for tweets
end_date = datetime.now()
start_date = end_date - timedelta(days=180)

# Extract tweets with the hashtag and within the specified time interval
tweets = tweepy.Cursor(api.search_tweets,
              q=hashtag,
              lang="en",
              tweet_mode='extended',
              since_id=start_date,
              max_id=end_date).items(100)

# Create a list to store the tweets
tweet_list = []

# Iterate over the tweets and store the relevant information in the list
for tweet in tweets:
    tweet_list.append({
        "created_at": tweet.created_at,
        "text": tweet.full_text,
        "user_id": tweet.user.id,
        "user_name": tweet.user.screen_name,
        "user_location": tweet.user.location
    })
# Print the number of tweets extracted
print("Number of tweets extracted:", len(tweet_list))
# Convert the list of tweets into a Pandas DataFrame
df5 = pd.DataFrame(tweet_list)

# Organize the format of the DataFrame
#df5 = df5[["created_at", "text", "user_id", "user_name", "user_location"]]

# Print the DataFrame
print(df5)
```

```python
import requests
import pandas as pd

# Define the API endpoint
endpoint = "https://api.nftmarketplace.com/v1/nfts/<nft_id>"

# Send a request to the API to get the NFT data
response = requests.get(endpoint)

# Extract the NFT data from the response
nft_data = response.json()

# Store the NFT data in a pandas DataFrame
df = pd.DataFrame(nft_data)

# Plot the price data over time
df.plot(x='timestamp', y='price')
```
